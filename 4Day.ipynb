```python
class Calculator:
    def setData(self, first, second): # setData()는 멤버필드를 초기화 하기 위해서
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second
```

```python
a = Calculator() # 객체를 생성
a.setData(10, 5) # 멤버필드 초기화
a.add()
print(a.result)
a.div()
print(a.result)
```

```python
class Calculator: 
    # 생성자
    def __init__(self, first, second): # setData()는 멤버필드를 초기화 하기 위해서
        self.first = first
        self.second = second
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second
```

```python
a = Calculator(10, 5) #객체 생성시에  __init__가 자동으로 실행이 된다.
a.add()
print(a.result)
a.div()
print(a.result)
```

```python
class FourCal(Calculator): # 상속 : Calculator:부모 , FourCal:자식
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
    def div(self): # 재정의 : override : 부모한테 받은 메서드를 자식이 다시 정의 하는 것
        if self.second == 0:
            self.result = 0
        else :
            self.result = self.first / self.second
```

```python
a = FourCal(5, 0) 
a.div() # 5 / 0 : 불능
print(a.result)
```

```python
# 스크래핑
import requests 
# 크롤링
from bs4 import BeautifulSoup
# 데이터분석
import pandas
# 시각화
import matplotlib as mpl 
import matplotlib.pyplot as plt
# 스크래핑
url = "https://www.weather.go.kr/w/observation/land/city-obs.do"
source  = requests.get(url) 

# 크롤링
soup = BeautifulSoup(source.content, 'html.parser')
table = soup.find("table",{"class":"table-col"})
#print(table)
data =[] #  크롤링 한데이터 저장
for tr in table.find_all('tr'):
    ths = list(tr.find_all('th'))
    tds = list(tr.find_all('td'))
    for th in ths:
        if th.find('a'):
            point = th.find('a').string
            temp = tds[4].string
            humidity = tds[8].string
            data.append([point, temp, humidity])
# 데이터를 분석
# 데이터 저장
with open("weather.csv" , "w" , encoding="utf-8") as f:
    f.write('지역,온도,습도\n')
    for l in data:
        f.write(f"{l[0]},{l[1]},{l[2]}\n")
#테이터 처리
df = pandas.read_csv("weather.csv",  encoding='utf-8', index_col='지역')
#print(df)
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
#print(city_df)
#시각화
ax = city_df.plot(kind='bar', title='weather', figsize=(12,4))
plt.show()
```
