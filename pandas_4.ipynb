```python
a = pd.Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan],
              index=["f", "e", "d", "c", "b", "a"])
b = pd.Series([0., np.nan, 2., np.nan, np.nan, 5.],
              index=["a", "b", "c", "d", "e", "f"])
print(a)
print(b)
np.where(pd.isna(a), b, a)
# np.where(condition, x, y)
# condition: 조건 (Boolean 배열)
# x: 조건이 True일 때 선택될 값
# y: 조건이 False일 때 선택될 값
```

```python
a.combine_first(b)
# a의 값 중에서 NaN인 항목을 b에서 동일한 인덱스의 값으로 채워 줍니다.
# a에 값이 있으면 그대로 유지, a가 NaN이면 b에서 채움
```

```python
df1 = pd.DataFrame({"a": [1., np.nan, 5., np.nan],
                    "b": [np.nan, 2., np.nan, 6.],
                    "c": range(2, 18, 4)})
df2 = pd.DataFrame({"a": [5., 4., np.nan, 3., 7.],
                    "b": [np.nan, 3., 4., 6., 8.]})
print(df1)
print(df2)
df1.combine_first(df2)
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/soldesk20250614/refs/heads/main/examples/macrodata.csv")
print(data)
data = data.loc[:, ["year", "quarter", "realgdp", "infl", "unemp"]]
data.head()
```

```python
print(data)
# pd.PeriodIndex는 연, 분기, 월, 주 등의 **정해진 주기(period)**를 인덱스로 갖는 시계열 데이터를 표현하는 Pandas의 인덱스 객체
periods = pd.PeriodIndex(year=data.pop("year"),
                         quarter=data.pop("quarter"),
                         name="date")
print(data)
print(periods)
data.index = periods.to_timestamp("D")
data.head()
```
